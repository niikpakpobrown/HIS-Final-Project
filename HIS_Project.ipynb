## Goal of the Project

The aim of this project is to develop a Decision Support System (DSS) that uses machine learning to predict the risk of dementia in patients based on their cognitive, genetic, and lifestyle profiles. The system will generate a personalized dementia risk index and provide evidence-based recommendations, such as cognitive training exercises, dietary changes, and physical activity plans, to delay disease progression. By integrating these features, the DSS will support early detection, patient monitoring, and informed decision-making, benefiting clinicians, caregivers, and patients.

The project will begin with data exploration and cleaning using Kaggle’s Alzheimer’s Prediction Dataset, which includes a diverse set of records from multiple countries. The system will align with best practices by incorporating Alzheimer’s prevention and treatment guidelines from trusted sources like the World Health Organization (WHO) and the National Institute on Aging (NIA).

This work will lay the foundation for building a robust machine learning model that can enhance the healthcare community's ability to identify and intervene in dementia cases early, improving the quality of life for at-risk individuals.

### DISCUSSION AND CONCLUSIONS
Our exploratory data analysis (EDA), machine learning (ML) classification, and correlation analyses provided valuable insights into the factors associated with Alzheimer’s Disease (AD) diagnosis, with social and cognitive function factors appearing to contribute most strongly. Key takeaways include:

Dataset distribution and initial EDA:

- The dataset largely is balanced for each risk factor (e.g., gender). The age range of participants for the study is between 69-90 years of age, and roughly 35% of participants are diagnosed with Alzheimer's. Initial exploration shows that sleep quality is slightly lower and BMI is slightly higher for AD patients. Family history of AD and cardiovascular disease also shows slight association with AD diagnosis, and higher education level is slightly correlated with lower rates of AD diagnosis.

ML Model Selection: 

- 6 ML classification models were trained, including Logistic Regression, KNN, Naive Bayesian, Random Forest, Decision Tree, and Gradient Boosting. Gradient Boosting yielded the highest accuracy, with 92.16%.


Medical History, Biomarkers, and Lifestyle Factors (H1):

- The EDA, ML analysis, and correlation studies revealed no significant relationship between AD diagnosis and individual medical history, biomarkers, or lifestyle factors (-0.056 < r < 0.054). This suggests that these factors may not independently explain AD risk or diagnosis, aligning with the idea that AD is a multifactorial disease.

Cognitive and Functional Assessments (H2):

- The most predictive features in the ML model were MMSE scores, ADL measures, functional assessments, memory complaints, and behavioral problems (0.101 - 0.235). These features also showed the highest correlations with AD diagnosis (0.24 < |r| < 0.37), highlighting their importance in identifying AD.

Genetic and Risk Factor Associations (H3):

- While the APOE gene is strongly associated with nearly all risk factors in our dataset, chi-square analysis (chi2 = 952.294, p = 1.0) suggests that APOE alone may not explain these associations. Even when APOE is excluded from the gene dataset, chi-square analysis (chi2 = 1074.06, p = 1.0) shows no evidence specific interactions between a single gene and a single risk factor. This points to the likelihood of gene-gene interactions or combinations of genes acting synergistically to influence AD risk factors.
  
Speculation and Broader Implications
While the analyses did not yield statistically significant conclusions in several areas, they suggest intriguing possibilities:

- Social Stimulation and Cognitive Perception:
    - Social stimulation and an individual's perception of their cognitive functionality may play critical roles in delaying the onset of AD. These psychosocial factors could complement biological and medical interventions in AD prevention strategies.

- Gene Synergism and Risk Factor Interaction:
    - The chi-square analysis hints at the possibility of epistasis (gene-gene interactions) influencing risk factor development. Further investigation into the interplay of APOE with other genes could deepen our understanding of its role in AD.

Future Directions

To perform deeper and more conclusive analyses with the gene interaction data, the following approaches should be considered:

Clustering and Subgroup Analysis:

Epistasis Analysis:
- Investigate the interaction effects between genes using advanced genetic analysis techniques to better understand how combinations of genes contribute to specific risk factors and overall AD susceptibility.

Conclusion

While no single variable or factor appears to drive AD diagnosis independently, our findings underscore the importance of multi-dimensional approaches in AD research. This study highlights the need for continued exploration into the synergistic effects of genetics, cognitive functionality, and social factors in the development and prevention of Alzheimer’s Disease. ## Plots of diagnosis agianst various conditions

```python

# General function to plot a variable against Alzheimer's Diagnosis
def plot_variable_against_diagnosis(df, variable):
    # Mapping for the columns with values that need to be renamed for clarity
    if variable == 'Smoking Status':
        mapping = {'Never': 'No Smoking', 'Former': 'Former Smoking', 'Current': 'Current Smoking'}
    elif variable == 'Family History of Alzheimer’s':
        mapping = {'No': 'No Family History', 'Yes': 'Family History of Alzheimer’s'}
    elif variable == 'Hypertension':
        mapping = {'No': 'No Hypertension', 'Yes': 'Hypertension'}
    elif variable == 'Diabetes':
        mapping = {'No': 'No Diabetes', 'Yes': 'Diabetes'}
    elif variable == 'Depression Level':
        mapping = {'Low': 'Low Depression', 'Medium': 'Medium Depression', 'High': 'High Depression'}
    elif variable == 'Cholesterol Level':
        mapping = {'Normal': 'Normal Cholesterol', 'High': 'High Cholesterol'}
    elif variable == 'Alcohol Consumption':
        mapping = {'Never': 'Never Consumed Alcohol', 'Occasionally': 'Occasionally Consumed Alcohol', 'Regularly': 'Regularly Consumed Alcohol'}
    elif variable == 'Physical Activity Level':
        mapping = {'Low': 'Low Physical Activity', 'Medium': 'Medium Physical Activity', 'High': 'High Physical Activity'}
    elif variable == 'Social Engagement Level':
        mapping = {'Low': 'Low Social Engagement', 'Medium': 'Medium Social Engagement', 'High': 'High Social Engagement'}
    elif variable == 'Income Level':
        mapping = {'Low': 'Low Income', 'Medium': 'Medium Income', 'High': 'High Income'}
    elif variable == 'Stress Levels':
        mapping = {'Low': 'Low Stress', 'Medium': 'Medium Stress', 'High': 'High Stress'}
    elif variable == 'Genetic Risk Factor (APOE-ε4 allele)':
        mapping = {'No': 'No Genetic Risk', 'Yes': 'Genetic Risk (APOE-ε4)'}
    else:
        mapping = None  # No mapping needed for numerical columns like 'Age' or 'BMI'

    # Apply the mapping if required
    if mapping:
        df[variable] = df[variable].map(mapping)

    # Check the unique values of 'Alzheimer's Diagnosis' and determine the palette
    if df["Alzheimer's Diagnosis"].dtype == 'object':  # If categorical (Yes/No)
        palette = {'No': '#CEDFEF', 'Yes': '#92C2DD'}
    else:  # If numeric (0/1)
        palette = {0: '#CEDFEF', 1: '#92C2DD'}

    # Count the occurrences of the variable values for each diagnosis group
    variable_counts = df.groupby([variable, "Alzheimer's Diagnosis"]).size().reset_index(name='Count')

    # Create the bar plot
    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=variable_counts,
        x=variable,
        y='Count',
        hue="Alzheimer's Diagnosis",  # Use diagnosis as hue
        palette=palette  # Correct palette based on diagnosis values
    )

    # Add bar labels
    for container in plt.gca().containers:
        plt.bar_label(container, fmt='%.1f')

    # Title and labels
    plt.title(f"{variable} and Alzheimer's Diagnosis")
    plt.xlabel(variable)
    plt.ylabel("Count")

    # Custom legend for Alzheimer's Diagnosis
    No_patch = mpatches.Patch(color='#CEDFEF', label="No Alzheimer's")
    Yes_patch = mpatches.Patch(color='#92C2DD', label="Alzheimer's")
    plt.legend(handles=[No_patch, Yes_patch], title="Diagnosis")

    # Show plot
    plt.show()

# List of variables to plot
variables = [
    'Smoking Status', 
    'Family History of Alzheimer’s', 
    'Hypertension', 
    'Diabetes', 
    'Depression Level', 
    'Cholesterol Level', 
    'Alcohol Consumption',  
    'Physical Activity Level', 
    'Social Engagement Level', 
    'Income Level', 
    'Stress Levels', 
    'Genetic Risk Factor (APOE-ε4 allele)'
]

# Plot each variable against Alzheimer's diagnosis
for var in variables:
    plot_variable_against_diagnosis(df, var)
```

### Bar plot for analyzing if there is relationship between educational level and AD.


```python

# Creating bins for Education Level
bins = [0, 8, 12, 19]  # Defining the bins for Low, Medium, High education
labels = ['Low Education', 'Medium Education', 'High Education']  # Labels for each bin
df['Education Category'] = pd.cut(df['Education Level'], bins=bins, labels=labels, right=True)

# Count the occurrences of education categories for each Alzheimer's diagnosis group
education_counts = df.groupby(['Education Category', "Alzheimer's Diagnosis"]).size().reset_index(name='Count')

# Calculate the total count for each Education Category
education_total = education_counts.groupby('Education Category')['Count'].transform('sum')

# Calculate the percentage for each group
education_counts['Percentage'] = (education_counts['Count'] / education_total) * 100

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=education_counts,
    x='Education Category',
    y='Percentage',
    hue="Alzheimer's Diagnosis",  # Use diagnosis as hue
    palette={0: '#CEDFEF', 1: '#92C2DD'}  # Color palette for No/Yes Alzheimer's
)

# Add bar labels showing percentages
for container in plt.gca().containers:
    plt.bar_label(container, fmt='%.1f%%')

# Title and labels
plt.title("Education Level and Alzheimer's Diagnosis (as Percentages)")
plt.xlabel('Education Level')
plt.ylabel("Percentage")

# Custom legend for Alzheimer's Diagnosis
No_patch = mpatches.Patch(color='#CEDFEF', label="No Alzheimer's")
Yes_patch = mpatches.Patch(color='#92C2DD', label="Alzheimer's")
plt.legend(handles=[No_patch, Yes_patch], title="Diagnosis")

# Show plot
plt.show()
```
